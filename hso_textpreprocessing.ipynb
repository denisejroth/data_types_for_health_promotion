## Text Preprocessing for Automated Text Analysis
This Jupyter Notebook covers fundamental text preprocessing techniques for automated text analysis.
It includes:
1. Cleaning text
2. Tokenization
3. Stopword removal
4. Stemming
5. Lemmatization
6. Bag-of-Words Representation

Each section contains explanations and code examples.

```python
# Import necessary libraries
import re
import nltk
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import PorterStemmer, WordNetLemmatizer
```

```python
# Download required NLTK datasets (only needed once)
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')
```

### **1. Sample Text Data**

We start with a few sample sentences to demonstrate the preprocessing techniques.

```python
data = [
    "This is a sample sentence! It contains punctuation & stopwords.",
    "Text preprocessing is essential for automated text analysis.",
    "Tokenization breaks text into words. Stemming reduces words to their root form.",
    "Lemmatization also reduces words but ensures they remain valid words."
]
```

### **2. Cleaning Text**

Cleaning text involves:
- Converting all text to lowercase.
- Removing special characters and punctuation.

```python
def clean_text(text):
    """Function to clean text by removing special characters and converting to lowercase."""
    text = text.lower()  # Convert to lowercase
    text = re.sub(r'[^a-zA-Z0-9\s]', '', text)  # Remove punctuation
    return text
```

```python
# Apply cleaning
data_cleaned = [clean_text(text) for text in data]
print("Cleaned Text:", data_cleaned)
```

### **3. Tokenization**

Tokenization splits text into individual words (tokens).

```python
data_tokenized = [word_tokenize(text) for text in data_cleaned]
print("Tokenized Text:", data_tokenized)
```

### **4. Removing Stopwords**

Stopwords are common words (e.g., 'the', 'is', 'and') that do not add much meaning to text analysis.

```python
stop_words = set(stopwords.words('english'))
data_no_stopwords = [[word for word in sentence if word not in stop_words] for sentence in data_tokenized]
print("Text without Stopwords:", data_no_stopwords)
```

### **5. Stemming**

Stemming reduces words to their root form using heuristic rules.
For example, 'running' becomes 'run'.

```python
stemmer = PorterStemmer()
data_stemmed = [[stemmer.stem(word) for word in sentence] for sentence in data_no_stopwords]
print("Stemmed Text:", data_stemmed)
```

### **6. Lemmatization**

Lemmatization also reduces words but ensures they remain valid words.
For example, 'running' becomes 'run', but 'better' remains 'better'.

```python
lemmatizer = WordNetLemmatizer()
data_lemmatized = [[lemmatizer.lemmatize(word) for word in sentence] for sentence in data_no_stopwords]
print("Lemmatized Text:", data_lemmatized)
```

### **7. Bag-of-Words Representation**

Bag-of-Words (BoW) converts text data into numerical form by counting word occurrences. This is our Document-Term Matrix (DTM).

```python

vectorizer = CountVectorizer()
X = vectorizer.fit_transform([" ".join(sentence) for sentence in data_lemmatized])
print("BoW Matrix:")
print(pd.DataFrame(X.toarray(), columns=vectorizer.get_feature_names_out()))
```
